```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'E:\Data Science Projects\Datasets\Updated_datasets\naukri_data.csv')
```

```python
data = df.copy()
```

```python
data.columns
```

```python
plt.figure(figsize=(14,6))
sns.stripplot(x = 'min_exp', y = 'min_pay', data = data)

## Now mapping Boxplot on the striplplot
sns.boxplot(x = 'min_exp', y = 'min_pay', data = data)
plt.title('Stripplots between min_exp and min_pay')
```

```python
plt.figure(figsize=(14,6))
sns.stripplot(x = 'max_exp', y = 'max_pay', data = data)

## Now mapping Boxplot on the striplplot
sns.boxplot(x = 'max_exp', y = 'max_pay', data = data)
plt.title('Stripplots between max_exp and max_pay')
```

## Find Median Max and Min pay in Different Industries

```python
data[['max_pay','industry']].groupby(['industry']).median().sort_values('max_pay',ascending = False).head(10)
```

```python
data[['max_pay','industry']].groupby(['industry']).median().sort_values('max_pay',ascending = False).head(10).plot(kind = 'bar',color = 'green')
```

```python
data[['min_pay','industry']].groupby(['industry']).median().sort_values('min_pay',ascending = False).head(10).plot(kind = 'bar',color = 'red')
```

## Which is my Top rated skills

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='skills',y = 'avg_payrate', data = data)
plt.xticks(rotation = 'vertical')
```

```python
data.groupby(['skills'])['avg_payrate'].median().reset_index().sort_values(by='avg_payrate',ascending = False).head(10).set_index('skills').plot(kind = 'bar',color = 'orange')
```

```python
def split_exp(exp):
    min_exp = exp.split('-')[0]
    max_exp = exp.split('-')[1]
    return min_exp,max_exp
```

```python
split_exp(value)
```

```python
len1 = []
for exp in df['experience'].dropna():
    if len(exp.split('-'))!=2:
        len1.append(exp)
```

```python
len1
```

```python
#Function using Exception Handling
def split_exp2(exp):
    try:
        if len(exp.split('-'))==2:
            min_exp = exp.split('-')[0]
            max_exp = exp.split('-')[1]
        return pd.Series([min_exp,max_exp])
    except:
        return pd.Series([np.nan,np.nan])
```

```python
df['experience'].apply(split_exp2)
```

```python
nm = pd.DataFrame(df['experience'].str.contains('Not Mentioned'))
```

```python
nm
```

```python
nm[nm['experience']==True].index
```

```python
df[['min_exp','max_exp']] = df['experience'].apply(split_exp2).rename(columns={0:'min_exp', 1:'max_exp'})
```

```python
df['max_exp'] = df['max_exp'].str.replace('yrs','')
```

```python
df['avg_expereince'] = (df['min_exp'] + df['max_exp'])/2
df['avg_payrate'] = (df['min_pay'] + df['max_pay'])/2
```

## Clean and Perform Feature Engineering on PostDate Feature 

```python
# Changing the data type of PostDate
df['postdate'] = pd.to_datetime(df['postdate'])
```

```python
df.dtypes
```

```python
# Defining own custom function
def fetch_dt_att(dataframe,feature):
    try:
        return pd.Series([dataframe[feature].dt.day,dataframe[feature].dt.month,dataframe[feature].dt.year])
    except:
        print('Data Type is not supported')
```

```python
df[['day', 'month', 'year']] = fetch_dt_att(df,'postdate')
```

```python
df.head(2)
```

```python
# Defining map function

def fetch_dt_att2(x):
    return [x.day,x.month,x.year]
```

```python
fe_date = pd.DataFrame(map(fetch_dt_att2,df['postdate'])).rename(columns={0:'day',1:'month',2:'year'})
```

```python
pd.concat((df,fe_date),axis=1)
```

```python
# Using date-time module

df['postdate'].dtype
```

```python
### datetime is callable as ‘dt’ (as imported) ,  dt.month is a attribute just to find month
### df is a object of dataframe..

df['year'] = df['postdate'].dt.year
df['month'] = df['postdate'].dt.month
df['day'] = df['postdate'].dt.day
```
